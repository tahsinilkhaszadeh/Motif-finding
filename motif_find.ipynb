```python
# -*- coding: utf-8 -*-
"""
Created on Sun Feb  6 01:36:26 2022

@author: tahsin - rastegar
"""

import random

def randomizedmotifsearch(lines): #Finds random l_mer

    sTable=[]
    motifs = []

    for n in lines:
       for i in range(20):
           i = random.randint(0, 365)
           lmer=n[i:i+35]
           motifs.append(lmer)
           sTable.append({
                "seq": lmer,
                "place": i,
            })

    """for i in range(len(sTable)):
        print("p",i+1,"  ",sTable[i]["seq"] ) """

    return  sTable
#**************************************
def crossover(lines,sTable): #random cut point crossover

    Table=[]

    for n in range(25):
       k = random.randint(2, 33)

       i = random.randint(1, len(sTable)-1)  #selecting parent
       j = random.randint(1, len(sTable)-1)

       p1 = sTable[i]["seq"]
       p2 = sTable[j]["seq"]

       #interchanging the genes
       o1=p1[0:k+1]+p2[k+1:k+35]    #creating offspring
       o2=p2[0:k+1]+p1[k+1:k+35]

       Table.append({
            "seq": o1,
            "place": sTable[i]["place"],
        })

       Table.append({
            "seq": o2,
            "place": sTable[j]["place"],
        })

    """print('Offsprings '.center(60, '-'))
    for i in range(len(Table)):
        print("O",i+1,"  ",  Table[i]["seq"])"""

    sTable = Table

    return  Table

#************************************************************
# calculation of FS=Fitness score and TFS=Total fitness score
# Counts the Number of matching characters in a pair of string

def Score(seqs,lines):

    scoretable=[]
    #seqs= sTable[i]["seq"]
    FSlist=[]
    TFSlist=[]
    for line in lines:
        sj=[]
        for j in range(364):
            sj.append(line[j:j+35])

   #FS calculation with match scores -> match=1   mismatch=0
    FS=[]
    TFS=[]

    for j in range(len(seqs)):
        for a in range(len(sj)):
            s1= seqs[j]["seq"]
            s2= sj[a]

            sum1=0
            for i in range(35):
                if s1[i] == s2[i]:
                   sum1+= 1

            mutations= 35-sum1

            if (mutations <= 7):
                b=sum1/35
                FSlist.append(b)
            else:
                continue


        if (len(FSlist)==0) :
            return
        else:
            FS=max(FSlist)
            #TFS calculation with match scores
            TFSlist.append(FS)
            TFS=sum(TFSlist)
            scoretable.append({
                    "seq":seqs[j]["seq"],
                    "score":TFS
                    })


            """print("TFS= ",TFS)
            print("FS values: ",FSlist)
            for d in range(len(FSlist)):
               print("%6.4f "%FSlist[d])"""


            #finds max score to select motif
            m=scoretable[0]["score"]
            t=0
            for i in range(len(scoretable)):
                if (scoretable[i]["score"]> m):
                    m=scoretable[i]["score"]
                    t=i

            print("Potential motif : ",scoretable[t]["seq"] ,
                  "\tscore= ",scoretable[t]["score"]," \n")

            return scoretable[t]

#**********************************************
def motiffind(motif):

    import difflib

    match_strings=[]
    #places=[]
    table=[]
    for line in lines:
       sj=[]
       for j in range(364):
            sj.append(line[j:j+35])

       #n is the maximum number of close matches to return in each line
       match_strings=difflib.get_close_matches(motif,sj,n=1,cutoff=0.6)
       t=lines.index(line)
       #print("%d"%t,match_strings)
       for n in match_strings:
           place = line.find(n)

           table.append({
                    "seq": n,
                    "line" : t,
                    "place": place
                    })

    #print(table)
    for t in range(len(lines)):

      #return positions starting from 1
        print(" line: ",table[t]["line"]+1,"\t\tplace:",table[t]["place"]+1)
    return


#**********************************************
if __name__ == "__main__":

    motif_lenght=35
    bestmotif=[]
    with open('motif_seq_main.txt') as f:
        lines = f.readlines()
    f.close()
    global iteration_number
    iteration_number=10
    print(" finding potential motifs ".center(60,"-"))
    for i in range(iteration_number):
        potentialmotif = None
        while (potentialmotif == None):
             seqs1 = randomizedmotifsearch(lines)
             seqs2 = crossover(lines,seqs1)
             seqs=seqs1+seqs2
             potentialmotif=Score(seqs,lines)

        bestmotif.append(potentialmotif)

    m = bestmotif[0]["score"]
    t=0
    for i in range(1,len(bestmotif)):
        #finds max score to select motif
        if (bestmotif[i]["score"] > m):
              m = bestmotif[i]["score"]
              t=i

    #print(*bestmotif)
    motif=({
            "seq"  : bestmotif[t]["seq"],
            "score": bestmotif[t]["score"],
             })

    print(" final result ".center(60,"-"))
    print("best motif is:", motif["seq"],"\t score: ",motif["score"])

    print("")
    print(" places of best motif ".center(60,"-"),"\n")
    string=motif["seq"]
    motiffind(string)




```
